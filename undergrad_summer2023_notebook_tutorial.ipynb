You can call external routines just like in a python program.  Here I use the "numpy" library, which is a large library of numerical functions.  

```python
import numpy as np
```

The "array" data type in Numpy is a data structure that can store multi-dimensional lists of data, e.g. 2D arrays.  The following commands make a set of 1D arrays filled with different values.

```python
#fill with zeros
c = np.zeros(20)
print('an array of zeros',c)

#fill with integers
xint = np.arange(0,20,1)
print('an array of integers', xint)

#fill with sequential real numbers
xreal = np.arange(0,20,0.5)
print('an array of real numbers', xreal)

#the default for the "arange" is to fill with integers
x = np.arange(20)
print('the default output of arange', x)
```

### You can also make plots directly in a notebook

```python
y = 0.05 * xreal**2
y
```

```python
from matplotlib import pyplot as plt
#plot the data with two different point types
plt.plot(xreal, y,'b-',label = 'data line')
plt.plot(xreal, y,'ro',label = 'data points')

#make plot labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
```

you can make a series of plots that are displayed in a scroll window

```python
#create an array of multiplicative factors
a = np.arange(15)*0.01 + 0.01

for i in range(len(a)):
    #create an array of y values using each of the multiplicative factors
    #calculated above
    ya = a[i] * xreal**2
    
    #plot the data with two different point types
    plt.plot(xreal,ya,'b-',label = 'data line')
    plt.plot(xreal,ya,'ro',label = 'data points')

    #make plot labels
    plt.xlabel('x')
    plt.ylabel('y')
    
    #this creates a label string which converts the scale in the polynomial to a 
    #string with a formating statement to control the precision of the number
    scalestr = str("%1.2f" % a[i])
    #this makes a string by joining three different strings with a "+".
    #note that the last compnent has LaTeX in the string
    labstr = 'y = ' + scalestr + 'x$^2$'
    plt.title(labstr)
    
    #set the axis limits to be the same for each plot
    plt.xlim(-1,19)
    plt.ylim(-1,19)

    #putting this in after every plot makes a new plot
    #see what happens if you comment one or both of these both out.
    plt.legend(loc='upper left')
    plt.show()
```

# Best practices and pitfalls

*1. The order of cell execution matters.  All variable assignments are stored in memory and are available to any commands that are executed.  You need to be careful that you execute all cells in such an order so that variables are properly populated.  For example, imagine you execute the following cells*

```python
z = 3.0
```

```python
yy = z * 3
print(yy)
```

```python
z = 5.0
```

And then you go back to the middle block and execute it again.  It will use the last declaration of z (z=5.0) and not the first (z=3.0).  This will result in a different value for "yy"

*2. Use the markdown capability to describe what you have been doing, as I've been doing this whole time.  This is a strength of Jupyter Notebooks.  The ability to make a living notebook that seamlessly interleaves text, plots, and executable code.*

*3. Use the markdown capability to make figure captions.  For example*

```python
xreal = np.arange(0,20,0.5)
y = 0.2 * xreal + 2

#plot the data with two different point types
plt.plot(xreal, y,'b-',label = 'straight line')

#set the axis limits to be the same for each plot
plt.xlim(-1,19)
plt.ylim(-1,19)

#make plot labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
```

**caption**: This figure shows a line.  It's a very straight line.  

*4. When possible, don't use notebooks to develop code.  Instead develop your code outside of the notebook and implement it within the notebook using function calls to your code modules.  This will make your notebook much more readable.*

*4. Use notebooks to document your research as if you would be using a normal lab notebook.  When used with a version control system like **Github**, this allows you to keep the history of all of your edits.  This pairing makes python notebooks a way to document the path of your research.*

*5. A well put together Jupyter Notebook can serve as a tutorial or a way of documenting code and research to other members of your research community."*
